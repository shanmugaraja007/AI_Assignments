```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
def duplicate(input_list):
    return list(set([x for x in input_list if input_list.count(x) > 1]))
```

```python
data=pd.read_csv("50_Startups.csv")
data
```

```python
data=pd.get_dummies(data)
data
```

```python
data.columns
```

```python
independent=data[['R&D Spend', 'Administration', 'Marketing Spend',
       'State_California', 'State_Florida', 'State_New York']]
independent
```

```python
dependent=data[['Profit']]
dependent
```

```python
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
parameter=pd.read_csv("Parameter_Tune.csv")
parameter
```

```python
parameter.columns
```

```python
criterions=parameter['criterion'].tolist()
criterion =duplicate(parameter['criterion'].tolist())
criterion
```

```python
splitters=parameter['splitter'].tolist()
splitter=duplicate(parameter['splitter'].tolist())
splitter
```

```python
max_feature=parameter['max_features'].tolist()
max_features=duplicate(parameter['max_features'].tolist())
max_features
```

```python
r2=[]
for x in criterion:
    for y in splitter:
        for z in max_features:
                    regressor=DecisionTreeRegressor(criterion=x,splitter=y,max_features=z)
                    regressor.fit(x_train,y_train)
                    y_predict=regressor.predict(x_test)
                    r_score=r2_score(y_test,y_predict)
                    r2.append(r_score)
r2
```

```python
Dict={"criterion":criterions,"splitter":splitters,"max_features":max_feature,"R2_Score":r2}
Para_result = pd.DataFrame(Dict)
Para_result.to_csv('Parameter_Tune_Result.csv')
Para_result
```
